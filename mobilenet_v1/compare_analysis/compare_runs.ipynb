```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
with open("../baseline_build/verification/total_runtime_metrics.json",'r') as f:
    data = json.load(f)
df = pd.DataFrame.from_dict(data)
fpga_metrics = df[df.columns[0:2]]
fpga_metrics.to_csv("fpga_metrics.csv")
```

```python
df1 = pd.read_csv("../baseline_build/report/runtime_metrics.csv",index_col="Unnamed: 0")
df2 = pd.read_csv("../half_baseline_build/report/runtime_metrics.csv", index_col="Unnamed: 0")
df3 = df1[["FPGA throughput metrics","FPGA_metrics"]].append(df2[["FPGA throughput metrics","FPGA_metrics"]])
rt = df2.columns[2:]
df3.T.plot(kind='bar')
df3
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.ylabel("Runtime[ms]")
plt.title("Compare between first two runs")
plt.tight_layout()
plt.savefig("compare_fpga_metrics.svg")
# ax.set_yscale('log')

df4 = df1[rt].append(df2[rt])
df4.T.plot(kind='bar')
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.ylabel("Runtime[ms]")
plt.title("Compare between first two runs")
plt.tight_layout()
plt.savefig("compare_runtime.svg")
```
