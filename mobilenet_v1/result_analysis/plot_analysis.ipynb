```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("results_metrics.csv")
df = df.drop(columns = df.columns[-2:])
# df = df.set_index("Column1")
df

sns.lineplot(data=df)
plt.yscale("log")
plt.ylabel("runtime[ms]")
plt.xlabel("Run_id")
plt.title("Runtime metrics for split on the mobilenetv1 network")
plt.tight_layout()
plt.savefig("hw_run_graph.eps")
plt.savefig("hw_run_graph.png", dpi=400,facecolor = 'w')
plt.show()
```
